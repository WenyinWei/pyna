```python
import numpy as np 
```

```python
eigval_angle = 2.0
eigval1 = np.cos(eigval_angle) + np.sin(eigval_angle)*1j
eigval2 = np.cos(eigval_angle) - np.sin(eigval_angle)*1j
eigval = np.array([
    [ eigval1, 0      ],
    [ 0      , eigval2]
])
```

```python
eigvec = np.array([
    [ 560+ 200j, 560- 200j],
    [ -250+ 200j, -250- 200j]
])
eigvec1 = eigvec[:,0]
eigvec2 = eigvec[:,1]
```

```python
A = eigvec @ eigval @ np.linalg.inv(eigvec)
```

```python
init_RZ = np.array([1.0, 0.0])
```

```python
pts_RZ = [init_RZ, ]
for i in range(90):
    pts_RZ.append( A @ pts_RZ[-1] )
```

```python
import matplotlib.pyplot as plt
%matplotlib widget
fig, ax = plt.subplots(1, 1,)
ax.scatter(
    [pt_RZ[0] for pt_RZ in pts_RZ], 
    [pt_RZ[1] for pt_RZ in pts_RZ], )
ax.set_aspect(1)
```

```python
theta_btw_a_and_xaxispos = 2*np.pi/8
a_over_b = 1.5
theta_eig = 2*np.pi/8
def R_theta(theta:float):
    return np.array([
        [np.cos(theta),-np.sin(theta)],
        [np.sin(theta), np.cos(theta)],
    ])
def R_ell(theta_btw_a_and_xaxispos:float, theta_eig:float, a_over_b:float):
    R_ell_prime = np.array([
        [           np.cos(theta_eig),-a_over_b*np.sin(theta_eig) ],
        [1/a_over_b*np.sin(theta_eig),          np.cos(theta_eig) ]
    ])
    return R_theta(theta_btw_a_and_xaxispos) @ R_ell_prime @ R_theta(-theta_btw_a_and_xaxispos)
```

```python
init_RZ = np.array([1.0, 0.0])
A = R_ell(theta_btw_a_and_xaxispos, theta_eig, a_over_b)
```

```python
pts_RZ = [init_RZ, ]
for i in range(30):
    pts_RZ.append( A @ pts_RZ[-1] )
```

```python
import matplotlib.pyplot as plt
%matplotlib widget
fig, ax = plt.subplots(1, 1,)
ax.scatter(
    [pt_RZ[0] for pt_RZ in pts_RZ], 
    [pt_RZ[1] for pt_RZ in pts_RZ], )
ax.set_aspect(1)
```
