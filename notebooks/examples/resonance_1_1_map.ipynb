```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x, y = np.linspace(-2, 4, 2000), np.linspace(-2, 2, 2000)
xx, yy = np.meshgrid(x,y, indexing="ij")
```

```python
a0, b0 = 0.3, 0.3

beta1, beta2 = -0., -0.0

flowx_order0 = np.ones_like(xx) * (-1.0/2 * beta1)
flowy_order0 = np.ones_like(xx) * (1.0 * beta1)

flowx_order1 = yy
flowy_order1 = np.zeros_like(xx)

a00 = 1/20 * (2*b0-a0) * beta1**2 + 1/3*beta1*beta2
b00 = (a0/30 - 1/12*b0) * beta1**2 - 1/2 *beta1*beta2
flowx_order0_beta = a00
flowy_order0_beta = b00
a10, b10 = ((1/3)*b0 - 1/2*a0) * beta1,((2/3)*a0 - 1/2*b0) * beta1
a01, b01 = ((1/5)*a0 - 5/12*b0) * beta1 - 1/2 *beta2, ((1/2)*b0 - 1/6*a0) * beta1 + beta2
flowx_order1_beta = a10 * xx + a01*yy
flowy_order1_beta = b10 * xx + b01*yy
a20, b20 = -a0, 2*a0
a11, b11 = 2/3*a0 - 1/2*b0, b0-a0
a02, b02 = 2/3*b0 - 1/3*a0, 1/3*a0-b0
flowx_order2_beta = 1/2 * a20 * xx**2 + a11*xx*yy + 1/2 * a02 * yy**2
flowy_order2_beta = 1/2 * b20 * xx**2 + b11*xx*yy + 1/2 * b02 * yy**2

flowx = flowx_order0 + flowx_order1 + flowx_order0_beta + flowx_order1_beta + flowx_order2_beta
flowy = flowy_order0 + flowy_order1 + flowy_order0_beta + flowy_order1_beta + flowy_order2_beta
# flowx = flowx_order0 + flowx_order1 
# flowy = flowy_order0 + flowy_order1 
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,15))
ax.streamplot(x, y, flowx.T, flowy.T, color=np.sqrt(flowx.T**2+flowy.T**2), density=2.5)
```
