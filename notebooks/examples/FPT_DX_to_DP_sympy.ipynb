```python
import sympy
```

```python
Phi, R, Z, BR, BZ, BPhi = sympy.symbols("\phi, R, Z, B_{R}, B_{Z}, B_{\phi}", real=True)
```

```python
from sympy import sin, cos
```

```python
from sympy import Matrix
```

```python
m1 = Matrix([[1, 0,-R * BR/BPhi], [0, 1,-R * BZ/BPhi]])
```

```python
m2 = Matrix([
    [cos(Phi), 0, -R*sin(Phi)],
    [sin(Phi), 0,  R*cos(Phi)],
    [0, 1, 0          ]
])
```

```python
DX_11, DX_12, DX_13,DX_21,DX_22, DX_23,DX_31,DX_32,DX_33 =sympy.symbols("DX_{11}, DX_{12}, DX_{13},DX_{21},DX_{22}, DX_{23},DX_{31},DX_{32},DX_{33}", real=True)
```

```python
DX =Matrix([
    [DX_11, DX_12, DX_13],
    [DX_21,DX_22, DX_23],
    [DX_31,DX_32,DX_33]])
```

```python
DP = m1*(m2**-1)*DX*m2
```

```python
DP.simplify()
```

```python
DP.subs(Phi, 0)
```
