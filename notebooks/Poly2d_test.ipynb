$\mathcal{F}: \vec{x} \mapsto \mathrm{A}\vec{x} + \vec{F}^{(2)}(\vec{x}) + \vec{F}^{(3)}(\vec{x}) + \cdots, \vec{x}\in\mathbb{R}^{n}$

We wish to find a coordinate transform to diagonalize the map Jacobian

$\vec{x}=\mathrm{P}\vec{y}$

$\mathcal{F}: \mathrm{P}\vec{y} \mapsto \mathrm{A}\vec{x} + \vec{F}^{(2)}(\vec{x}) + \vec{F}^{(3)}(\vec{x}) + \cdots =\mathrm{A}\mathrm{P}\vec{y} + \vec{F}^{(2)}( \mathrm{P}\vec{y} ) + \vec{F}^{(3)}( \mathrm{P}\vec{y} ) + \cdots , \vec{x}\in\mathbb{R}^{n}$

Left multiplying $\mathrm{P}^{-1}$,

$\mathcal{F}: \vec{y} \mapsto \underbrace{\mathrm{P}^{-1}\mathrm{A}\mathrm{P} }_{\mathrm{\Lambda}(\mathrm{A}) } ~ \vec{y} + \mathrm{P}^{-1}\vec{F}^{(2)}( \mathrm{P}\vec{y} ) + \mathrm{P}^{-1}\vec{F}^{(3)}( \mathrm{P}\vec{y} ) + \cdots , \vec{x}\in\mathbb{R}^{n}$

In fact, $\mathrm{P} = \mathrm{V}$, the eigenvector matrix given by numpy.eig. Hence, $\mathrm{V\Lambda V}^{-1}=\mathrm{A}$.

```python
import numpy as np
from numpy import linalg as LA

Eigenvalues, Eigenvectors = LA.eig(np.array([
        [1, 3],
        [2, 5]
    ]))

Lambda = np.diag(Eigenvalues)


Eigenvectors @ Lambda @ LA.inv(Eigenvectors)
```

In code, the original map is changed by the transform from 

$ \mathcal{F} @ \vec{x}$  

to 

$ \mathrm{P} @ \mathcal{F} @ \mathrm{P}^{-1} @ \vec{x}$.
