```python
# Function that computes Fibonacci
# numbers with lru_cache
from functools import lru_cache
class Num:
    def __init__(self, n):
        self._n = n
    @property
    def n(self):
        return self._n
    
    @lru_cache(maxsize = 128)
    def fib_with_cache(N):
        if N.n < 2:
            return N.n
        return Num(N.n-1).fib_with_cache() + Num(N.n-2).fib_with_cache()
```

```python
a = Num(5)
```

```python
Num(50).fib_with_cache()
```

```python

# Function that computes Fibonacci
# numbers with lru_cache
@lru_cache(maxsize = 128)
def fib_with_cache(n):
    if n < 2:
        return n
    return fib_with_cache(n-1) + fib_with_cache(n-2)
```
