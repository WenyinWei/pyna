# Validation: chirikov_overlap & chaotic_boundary_estimate


```python
import numpy as np
from pyna.topo.chaos import chirikov_overlap, chaotic_boundary_estimate
```

## 1. Chirikov overlap criterion


```python
widths = np.array([0.05, 0.08, 0.04])
positions = np.array([0.3, 0.45, 0.65])
sigma = chirikov_overlap(widths, positions)
print('sigma =', sigma)
# sigma > 1 => overlap (chaos), sigma < 1 => separated islands
assert sigma.shape == (2,), 'Expected shape (N-1,)'
print('Test passed: sigma =', sigma)
```

## 2. chaotic_boundary_estimate on synthetic FTLE


```python
# Build a synthetic FTLE: high values in a circular region
nR, nZ = 50, 60
R = np.linspace(1.0, 2.5, nR)
Z = np.linspace(-0.5, 0.5, nZ)
R_grid, Z_grid = np.meshgrid(R, Z, indexing='ij')

# Chaotic blob centred at (R=1.75, Z=0)
ftle = np.exp(-((R_grid - 1.75)**2 + Z_grid**2) / 0.05)

R_bdy, Z_bdy = chaotic_boundary_estimate(ftle, R_grid, Z_grid, threshold_percentile=85.0)
print(f'Boundary points: {len(R_bdy)}')
assert len(R_bdy) > 0, 'Expected non-empty boundary'
print('Test passed!')
```

```python
try:
    import matplotlib.pyplot as plt
    fig, ax = plt.subplots()
    ax.pcolormesh(R_grid, Z_grid, ftle, cmap='inferno')
    ax.plot(R_bdy, Z_bdy, 'w.', ms=2, label='boundary')
    ax.set_xlabel('R'); ax.set_ylabel('Z'); ax.legend()
    plt.title('Synthetic FTLE with estimated chaotic boundary')
    plt.show()
except ImportError:
    print('matplotlib not available, skipping plot')
```
