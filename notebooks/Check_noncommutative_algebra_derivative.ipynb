```python
from sympy import Array, tensorcontraction, symbols, tensorproduct
t = symbols("t")

A1 = Array([
    [2, 1+t],
    [-5+3*t, 5+t],
])
A2 = Array([
    [
        [3, 2*t],
        [0, 4*t],
    ],
    [
        [0, 3*t],
        [-2+2*t, 5*t],
    ],
])
A3 = Array([
    [2+t, 1],
    [0, 5]
])

A1A2 = tensorcontraction(
    tensorproduct(A1,A2), (1, 2))
A1A2A3 = tensorcontraction(
    tensorproduct(A1A2,A3), (2, 3))
```
